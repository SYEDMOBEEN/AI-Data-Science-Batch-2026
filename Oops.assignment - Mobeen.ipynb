1. Create a class 'BankAccount' with attributes account holder name, account number, and
balance. Add methods for deposit and withdrawal. Implement logic to prevent withdrawal if balance
is insufficient.

```python
class BankAccount:
    def __init__(self, account_holder: str, account_number: str, initial_balance: float = 0.0):
    
        self.account_holder = account_holder
        self.account_number = account_number
        self.balance = initial_balance

    def deposit(self, amount: float):
        
        if amount > 0:
            self.balance += amount
            print(f"Success: ${amount: .2f} deposited into account {self.account_number}.")
        else:
            print("Error: Deposit amount must be greater than zero.")

    def withdraw(self, amount: float):
        
        if amount <= 0:
            print("Error: Withdrawal amount must be greater than zero.")
        elif amount > self.balance:
            print(f"Transaction Denied: Insufficient balance. Remaining funds: ${self.balance:.2f}")
        else:
            self.balance -= amount
            print(f"Success: ${amount:.2f} withdrawn from account {self.account_number}.")

    def display_balance(self):
        print(f"Account Details -> Holder: {self.account_holder} | Account: {self.account_number} | Current Balance: ${self.balance:.2f}")
```

2. Design a class 'Student' with attributes name and marks. Create a method to calculate grade
based on marks using conditions (A, B, C, Fail).

```python
class Student:
    def __init__(self, name, marks):
        
        self.name = name
        self.marks = marks

    def calculate_grade(self):
        
        if self.marks >= 85:
            return "A"
        elif self.marks >= 70:
            return "B"
        elif self.marks >= 50:
            return "C"
        else:
            return "Fail"

student1 = Student("Alice", 92)
student2 = Student("Bob", 74)
student3 = Student("Charlie", 55)
student4 = Student("David", 42)

students = [student1, student2, student3, student4]
for student in students:
    print(f"Student: {student.name} | Marks: {student.marks} | Grade: {student.calculate_grade()}"
```

3. Create a class 'ElectricityBill' for a KE-electric consumer. Include units consumed and calculate
bill based on slab rates (e.g., different price per unit for different ranges).

```python
class ElectricityBill:
    def __init__(self, consumer_name, sanctioned_load_kw=1.0):
        self.consumer_name = consumer_name
        self.sanctioned_load_kw = sanctioned_load_kw
        self.fixed_charge_rate = 275.0 
        self.gst_percentage = 0.18

    def calculate_bill(self, units_consumed):
        remaining_units = units_consumed
        base_bill = 0.0

        slabs = [
            (50, 3.95), 
            (50, 7.74), 
            (100, 10.29),
            (100, 12.55),
            (100, 17.38),
            (100, 18.84)
        ]
        for slab_limit, rate in slabs:
            if remaining_units <= 0:
                break
            units_in_slab = min(remaining_units, slab_limit)
            base_bill += units_in_slab * rate
            remaining_units -= units_in_slab

        if remaining_units > 0:
            base_bill += remaining_units * 20.40 

        fixed_charges = self.sanctioned_load_kw * self.fixed_charge_rate
        sub_total = base_bill + fixed_charges
        gst_amount = sub_total * self.gst_percentage
        
        tv_fee = 35.0 

        total_amount = sub_total + gst_amount + tv_fee

        return {
            "Units Consumed": units_consumed,
            "Base Charges (PKR)": round(base_bill, 2),
            "Fixed Charges (PKR)": round(fixed_charges, 2),
            "GST (18% in PKR)": round(gst_amount, 2),
            "TV Fee (PKR)": tv_fee,
            "Total Payable (PKR)": round(total_amount, 2)
        }
consumer = ElectricityBill(consumer_name="Ali Khan", sanctioned_load_kw=3.0)
bill_details = consumer.calculate_bill(250)

for key, value in bill_details.items():
    print(f"{key}: {value}")
```

4. Design a class 'Vehicle' and create subclasses 'Car' and 'Bike'. Override a method
'start_engine()' differently in each subclass to demonstrate polymorphism.

```python
class Vehicle:
    def start_engine(self):
        return "Starting a generic vehicle engine."

class Car(Vehicle):
    def start_engine(self):
        return "The car engine purrs to life with a key turn or push button."

class Bike(Vehicle):
    def start_engine(self):
        return "The bike engine roars to life after a kickstart or electric ignition."

def initiate_trip(vehicle_instance: Vehicle):
    print(vehicle_instance.start_engine())

generic_vehicle = Vehicle()
my_car = Car()
my_bike = Bike()

initiate_trip(generic_vehicle)  
initiate_trip(my_car)           
initiate_trip(my_bike)    
```

5. Create a class 'Employee' with basic salary. Add a method to calculate total salary including
bonus (10%) and tax deduction (5%).

```python
class Employee:
    def __init__(self, name, basic_salary):
        self.name = name
        self.basic_salary = basic_salary

    def calculate_total_salary(self):
        bonus = self.basic_salary * 0.10
        tax = self.basic_salary * 0.05
        total_salary = self.basic_salary + bonus - tax
        return total_salary

emp = Employee("Alice", 50000)
print(f"Employee: {emp.name}")
print(f"Basic Salary: ${emp.basic_salary}")
print(f"Total Salary (with 10% bonus and 5% tax): ${emp.calculate_total_salary()}")
```

6. Design an abstract class 'Shape' with a method 'area()'. Create subclasses 'Circle' and
'Rectangle' and implement area calculation logic.

```python
from abc import ABC, abstractmethod
import math


class Shape(ABC):
    
    @abstractmethod
    def area(self):
        
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
    
        return math.pi * (self.radius ** 2)

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
    
        return self.width * self.height

if __name__ == "__main__":

    my_circle = Circle(radius=5)
    my_rectangle = Rectangle(width=4, height=6)

    print(f"Circle Area: {my_circle.area():.2f}")       
    print(f"Rectangle Area: {my_rectangle.area():.2f}") 
    
```

7. Create a class 'HospitalPatient' with attributes name, age, and disease. Add a method to
determine priority (Emergency if age > 60 or serious disease).

```python
class HospitalPatient:
    def __init__(self, name, age, disease):
        self.name = name
        self.age = age
        self.disease = disease

    def get_priority(self, serious_diseases):
        
        if self.age > 60 or self.disease in serious_diseases:
            return "Emergency"
        return "Regular"

serious_conditions = ["heart attack", "stroke", "severe trauma"]
patient1 = HospitalPatient("John Doe", 65, "flu")
patient2 = HospitalPatient("Jane Smith", 45, "heart attack")

print(f"{patient1.name}: {patient1.get_priority(serious_conditions)}")
print(f"{patient2.name}: {patient2.get_priority(serious_conditions)}")
```

8. Build a class 'ShoppingCart' where user can add items with price. Implement logic to calculate
total bill and apply discount if total exceeds a certain amount.

```python
class ShoppingCart:
    def __init__(self, discount_threshold=100.0, discount_percent=10.0):
    
        self.items = [] 
        self.discount_threshold = discount_threshold
        self.discount_percent = discount_percent

    def add_item(self, item_name, price):
        """Adds an item and its price to the cart."""
        if price < 0:
            raise ValueError("Price cannot be negative.")
        self.items.append({"name": item_name, "price": float(price)})
        print(f"Added {item_name} (${price:.2f}) to your cart.")

    def calculate_subtotal(self):
        return sum(item["price"] for item in self.items)

    def calculate_total_bill(self):
        subtotal = self.calculate_subtotal()
    
        if subtotal > self.discount_threshold:
            discount_amount = subtotal * (self.discount_percent / 100)
            final_total = subtotal - discount_amount
            return final_total, discount_amount
        
        return subtotal, 0.0

    def print_receipt(self):
        if not self.items:
            print("\nYour shopping cart is empty.")
            return

        print("\n--- RECEIPT ---")
        for item in self.items:
            print(f"{item['name']}: ${item['price']:.2f}")
        print("-" * 15)
        
        subtotal = self.calculate_subtotal()
        final_total, discount_deducted = self.calculate_total_bill()
        
        print(f"Subtotal: ${subtotal:.2f}")
        if discount_deducted > 0:
            print(f"Discount Applied ({self.discount_percent}%): -${discount_deducted:.2f}")
        print(f"Total Bill: ${final_total:.2f}")
        print("----------------")
```

9. Create a class 'MobilePhone' with private attribute battery percentage. Use getter and setter
methods to update battery with validation (0–100 only).

```python
class MobilePhone:
    def __init__(self, battery_percentage=0):
        self.battery = battery_percentage  # Calls the setter for validation

    @property
    def battery(self):
        return self._battery

    @battery.setter
    def battery(self, value):
        if 0 <= value <= 100:
            self._battery = value
        else:
            raise ValueError("Battery percentage must be between 0 and 100.")

my_phone = MobilePhone(50)
print(f"Current Battery: {my_phone.battery}%")

my_phone.battery = 85
print(f"Updated Battery: {my_phone.battery}%")
```

10. Design a class 'Library' where users can borrow and return books. Add logic to check
availability before issuing a book.

```python
class Book:
    def __init__(self, isbn: str, title: str, author: str, total_copies: int = 1):
        self.isbn = isbn
        self.title = title
        self.author = author
        self.total_copies = total_copies
        # Initially, all copies are available for borrowing
        self.available_copies = total_copies

    def borrow_copy(self) -> bool:
        if self.available_copies > 0:
            self.available_copies -= 1
            return True
        return False

    def return_copy(self) -> bool:
        if self.available_copies < self.total_copies:
            self.available_copies += 1
            return True
        return False


class Library:
    def __init__(self):
        # Maps ISBN strings to Book objects for O(1) tracking lookups
        self.catalog = {}

    def add_book(self, book: Book):
        if book.isbn in self.catalog:
            self.catalog[book.isbn].total_copies += book.total_copies
            self.catalog[book.isbn].available_copies += book.total_copies
        else:
            self.catalog[book.isbn] = book
        print(f"Success: Added '{book.title}' to the library catalog.")

    def borrow_book(self, isbn: str, user_name: str) -> str:
        
        # Step 1: Check if the book exists in the library catalog
        if isbn not in self.catalog:
            return f"Error: Book with ISBN {isbn} does not belong to this library."
        
        book = self.catalog[isbn]
        
        # Step 2: Validate current physical copy availability
        if book.borrow_copy():
            return f"Success: '{book.title}' has been successfully issued to {user_name}. (Remaining copies: {book.available_copies})"
        else:
            return f"Unavailable: Sorry {user_name}, all copies of '{book.title}' are currently checked out."

    def return_book(self, isbn: str, user_name: str) -> str:
        
        if isbn not in self.catalog:
            return f"Error: Invalid return. ISBN {isbn} is not part of this library."
        
        book = self.catalog[isbn]
        if book.return_copy():
            return f"Success: {user_name} returned '{book.title}'. (Available copies: {book.available_copies})"
        else:
            return f"Error: All copies of '{book.title}' are already accounted for in the library."
```
