# User se marks lo aur grade do:

# 90+ → A
# 80+ → B
# 70+ → C
# 60+ → D
# warna → Fail

```python
Marks = int(input("Enter your marks: "))

if Marks >= 90:
    print("A")
elif Marks >= 80:
    print("B")
elif Marks >= 70:
    print("C")
elif Marks >= 60:
    print("D")
else:
    print("Fail")
```

# User se day number (1–7) lo:

# 1 → Monday
# 2 → Tuesday
## ...
# 7 → Sunday

```python
Day = input("Enter your day number between 1 to 7:")
if Day =="1":
    print("Monday")
elif Day =="2":
    print("Tuesday")
elif Day =="3":
    print("Wednesday")
elif Day =="4":
    print("Thursday")
elif Day =="5":
    print("Friday")
elif Day=="6":
    print("Saturday")
elif Day =="7":
    print("Sunday")
else:
    print("There is only seven days in a week")
```

# User se month number lo:

# 12,1,2 → Winter
# 3,4,5 → Spring
# 6,7,8 → Summer
# warna → Autumn

```python
month_number= input("Enter your month number:")
if month_number == "12" or month_number =="1" or month_number=="2":
    print("Winter")
elif month_number=="3" or month_number=='4' or month_number=="5":
    print("Spring")
elif month_number=="6" or month_number=="7" or month_number=="8":
    print("Summer")
else:
    print("Autumn")
```

# Simple calculator:
# User se operator lo (+, -, *, /):

# → addition
# → subtraction
# → multiplication
# / → division
# warna → "Invalid operator"

```python
num1 = int(input("Enter your number"))
Operator = input("Enter operator(+, -,*,/)")
num2 = int(input("Enter your second number"))
if Operator =="+":
    print(num1+num2)
elif Operator =="-":
    print(num1-num2)
elif Operator =="*":
    print(num1*num2)
elif Operator =="/":
    print(num1/num2)
else:
    print("Invalid Operator")
```
